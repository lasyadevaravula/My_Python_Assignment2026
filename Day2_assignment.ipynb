```python
# Q1.Check whether a person is eligible to vote or not.
age = int(input("enter your age: "))
if age >= 18:
    print("Eligible")
else:
    print("Not eligible")
```

```python
# Q2.Find the largest among three numbers.
a = 10 
b = 20
c = 30
if a>=b and a>=c:
    print("a is largest")
elif b>=a and b>=c:
    print("b is largest")
else:
    print("c is largest")
```

```python
# Q3.Write a program to check whether a year is a leap year.
year = int(input("enter a year: "))
if (year % 400 == 0):
    print("leap year")
else:
    print("not a leap year")
```

```python
# Q4.Create a simple calculator using if-else conditions.
a = int(input("enter first number: "))
b = int(input("enter second number: "))
operator = input("enter an operator (+, -, *, /): ")
if operator == "+":
    print(a + b)
elif operator == "-":
    print(a - b)
elif operator == "*":
    print(a * b)
elif operator == "/":
    print(a / b)
else:
    print("invalid operator")
```

```python
# Q5.Print the multiplication table of a given number. 
# (eg:5*1=5, 5*2=10 and so on till 10)
num = int(input("enter a number: "))
for i in range(1, 11):
    print(num, "*", i, "=", num * i)
```

```python
# Q6.Print all prime numbers between 1 and 100.
for num in range(1, 101):
    if num > 1:
        for i in range(2, num):
            if (num % i) == 0:
                break
        else:
            print(num)
```

```python
# Q7.Write a program to find the factorial of a number.
num = int(input("enter a number: "))
factorial = 1
for i in range(1, num + 1):
    factorial *= i
print("The factorial of", num, "is", factorial)
```

```python
# Q8.Reverse a given number using loops
num = int(input("enter a number: "))
reverse = 0
while num > 0:
    digit = num % 10
    reverse = reverse * 10 + digit
    num = num // 10
print("The reversed number is:", reverse)
```
