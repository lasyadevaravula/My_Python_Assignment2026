```python
# Q7.Create the following dataframe.
| Student | Marks | City      |
| ------- | ----- | --------- |
| A       | 85    | Hyderabad |
| B       | NaN   | Chennai   |
| C       | 92    | NaN       |
| D       | 78    | Bangalore |
| E       | NaN   | Mumbai    |
```

```python
# Q7.1 Find total missing values in each column.
import pandas as pd
import numpy as np
df = pd.DataFrame({
    'Students' : ['A', 'B', 'C', 'D', 'E'],
    'Marks' : [85, np.nan, 92, 78, np.nan],
    'City' : ['Hyderabad', 'Chennai', np.nan, 'Bangalore', 'Mumbai']
})
print(df.isnull().sum())
```

```python
# Q7.2 Display rows containing missing values.
print(df[df.isnull().any(axis=1)])
```

```python
# Q7.3 Replace missing marks with average marks.
average_marks = df['Marks'].mean()
df['Marks'].fillna(average_marks, inplace=True)
print(df)
```

```python
# Q7.4 Replace missing city values with "Unknown".
df['City'].fillna('Unknown', inplace=True)
print(df)
```

```python
# Q.7.5 Drop rows having any missing value.
df.dropna(inplace=True)
print(df)
```

```python
# Q7.6 Drop rows only if Marks is missing.
df.dropna(subset=['Marks'], inplace=True)
print(df)
```

```python
# 8. Create the following dataframe
| Order_ID | Product | Category    | Sales |
| -------- | ------- | ----------- | ----- |
| 1        | Laptop  | Electronics | 50000 |
| 2        | Mobile  | Electronics | 25000 |
| 3        | Chair   | Furniture   | 8000  |
| 4        | Table   | Furniture   | 12000 |
| 5        | Laptop  | Electronics | 55000 |
| 6        | Chair   | Furniture   | 9000  |
```

```python
# Q8.1 Find total sales by Category.
import pandas as pd
df = pd.DataFrame({
    'Order_ID' : [1, 2, 3, 4, 5, 6],
    'Product' : ['Laptop', 'Mobile', 'Chair', 'Table', 'Laptop', 'Chair'],
    'Category' : ['Electronics', 'Electronics', 'Furniture', 'Furniture', 'Electronics', 'Furniture'],
    'Sales' : [50000, 25000, 8000, 12000, 55000, 9000]
})
total_sales_by_category = df.groupby('Category')['Sales'].sum()
print(total_sales_by_category)
```

```python
# Q8.2 Find average sales by Category.
average_sales_by_category = df.groupby('Category')['Sales'].mean()
print(average_sales_by_category)
```

```python
# Q8.3 Find maximum sales by Category.
maximum_sales_by_category = df.groupby('Category')['Sales'].max()
print(maximum_sales_by_category)
```

```python
# Q8.4 Find minimum sales by Category.
minimum_sales_by_category = df.groupby('Category')['Sales'].min()
print(minimum_sales_by_category)
```

```python
# Q8.5 Find number of orders in each Category.
number_of_orders_by_category = df['Category'].value_counts()
print(number_of_orders_by_category)
```

```python
# Q8.6 Find total sales for each Product.
total_sales_by_product = df.groupby('Product')['Sales'].sum()
print(total_sales_by_product)
```

```python
# Q9. Create two Dataframes
| Emp_ID | Name  |
| ------ | ----- |
| 101    | Ravi  |
| 102    | Priya |
| 103    | Kiran |
| 104    | Sneha |

| Emp_ID | Department |
| ------ | ---------- |
| 101    | IT         |
| 102    | HR         |
| 103    | Finance    |
| 105    | Marketing  |
```

```python
# Q9.1 Perform Inner Join.
import pandas as pd
df1 = pd.DataFrame({
    'Emp_ID' : [101, 102, 103, 104],
    'Name' : ['Ravi', 'Priya', 'Kiran', 'Sneha']
})
df2 = pd.DataFrame({
    'Emp_ID' : [101, 102, 103, 105],
    'Department' : ['IT', 'HR', 'Finance', 'Marketing']
})
df_inner_join = pd.merge(df1, df2, on='Emp_ID')
print(df_inner_join)
```

```python
# Q9.2 Perform Left Join.
df_left_join = pd.merge(df1, df2, on='Emp_ID', how='left')
print(df_left_join)
```

```python
# Q9.3 Perform Right Join.
df_right_join = pd.merge(df1, df2, on='Emp_ID', how='right')
print(df_right_join)
```

```python
# Q9.4 Perform Outer Join.
df_outer_join = pd.merge(df1, df2, on='Emp_ID', how='outer')
print(df_outer_join)
```

```python
# Q9.5 Identify employees present in one DataFrame but not the other.
df_left_join = pd.merge(df1, df2, on='Emp_ID', how='left', indicator=True)
df_employees_not_in_df2 = df_left_join[df_left_join['_merge'] == 'left_only']
print(df_employees_not_in_df2)
```
