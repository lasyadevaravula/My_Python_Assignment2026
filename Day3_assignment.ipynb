```python
# Q1.Write a function to check whether a string is palindrome or not.
def palindrome(word):
    if word == word[::-1]:
        print("Palindrome")
    else:
        print("Not a Palindrome")
palindrome("madam")
palindrome("hello")
```

```python
# Q2.Write a function to find the second largest number in a list.
num = [10, 20, 30, 40, 50]
def second_largest(num):
    num.sort()
    print("Second largest num:", num[-2])
second_largest(num)
```

```python
# Q3.Write a function that accepts a list and returns only unique values.
def unique_values(data):
    unique = set(data)
    return unique
data = [1, 2, 3, 4, 5, 1, 2]
print(unique_values(data))
```

```python
# Q4.Use lambda to find the square of a number.
square = lambda x : x ** 2
print(square(5))
```

```python
# Q5.Use lambda with map() to convert a list of integers into squares.
numbers = [1, 2, 3, 4, 5]
squared_numbers = list(map(lambda x: x ** 2, numbers))
print(squared_numbers)
```

```python
# Q6.Use lambda with filter() to extract even numbers from a list.
numbers = [1, 2, 3, 4, 5, 6]
even_numbers = list(filter(lambda x: x % 2 == 0, numbers))
print(even_numbers)
```

```python
# Q7.Use lambda with sorted() to sort a list of tuples based on second value
data = [(1, 5), (2, 3), (3, 4)]
sorted_data = sorted(data, key = lambda x: x[1])
print(sorted_data)
```

```python
# Q8.Use lambda with reduce() to find the product of all numbers in a list.
from functools import reduce
num = [1,2,3,4,5]
result = reduce(lambda x, y: x * y, num)
print(result)
```

```python
# Q9.Use lambda to find the maximum between two numbers.
max_num = lambda a, b : x if a > b else b
print(max_num(10, 20))
```

```python
# Q10.Create Calculator using Functions
# Operations:Add, Subtract, Multiply, Divide.
# Use lambda for operations.
x = 10
y = 5
add = lambda x, y : x + y
subtract = lambda x, y : x - y
multiply = lambda x, y : x * y
divide = lambda x, y : x / y
print("Addition:", add(x, y))
print("Subtraction:", subtract(x, y))
print("Multiplication:", multiply(x, y))
print("Division:", divide(x, y))
```
