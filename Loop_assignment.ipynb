```python
# Exercise 1: Simple Age Check Write a program that checks if a person can get a driver's license (age >= 16).
user_age = int(input('Enter the age:'))
if user_age >= 16:
  print('You can get license')
else:
  print("You can't get license")
     
```

```python
# Exercise 2: Positive or Negative Write a program that checks if a number is positive, negative, or zero.

user_input = int(input("Enter a number to check if it's positive, negative or zero : "))
if user_input == 0:
  print('Zero')
elif user_input > 0:
  print('Positive Number')
elif user_input < 0 :
  print('Negative Number')
else:
  print('Not valid input! Enter number')
```

```python
# Exercise 3: Larger Number Write a program that compares two numbers and prints which one is larger.


num1 = int(input("Enter number 1 : "))
num2 = int(input("Enter number 2 : "))
if num1 > num2:
  print(f'{num1} is larger than {num2}')
elif num1 < num2:
  print(f'{num2} is larger than {num1}')
elif num1 == num2:
  print(f'{num1} is equal to {num2}')
else:
  print('Enter vaalid inputs!')
     
```

```python
marks = float(input("Enter student's marks (0-100): "))

if marks < 0 or marks > 100:
    print("Invalid marks! Please enter a value between 0 and 100.")
else:
    if marks >= 90:
        print("Grade A")
    elif marks >= 80:
        print("Grade B")
    elif marks >= 70:
        print("Grade C")
    elif marks >= 60:
        print("Grade D")
    else:
        print("Grade F")
```
