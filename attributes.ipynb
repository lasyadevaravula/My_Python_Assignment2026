## Attributes and methods
Attributes are the variables that belongs to objects or variables

```python
## instance attributes are the variables that are unique to each object.Every object has its own copy of instance attributes
```

```python
class Students:
    def __init__(self, name, roll_no):
        self.name = name
        self.roll_no = roll_no

student1 = Students("Rahul", 101)
student2 = Students("Priya", 102)

print(student1.name, student1.roll_no)
print(student2.name, student2.roll_no)
```

```python
print(student2.name)
print(student2.roll_no)
```

```python
class ClassName:
    class_attribute = "Example Value"

    def __init__(self):
        self.instance_attribute = "Instance Value"


obj = ClassName()

print(ClassName.class_attribute)
print(obj.instance_attribute)
```
