```python
# Q4. Create a list of 5 fruits:
fruits = ["Apple", "Banana", "Orange", "Grapes", "Mango"]
print("Fruits List:", fruits)
```

```python
# print first item
print("First Fruit:", fruits[0])
# print last item
print("Last Fruit:", fruits[-1])
```

```python
# add one more fruit to the list
fruits.append("Pineapple")
print("Updated Fruits List:", fruits)
```

```python
fruits.remove("Grapes")
print("Fruits List after removal:", fruits)
```

```python
# Reverse the list
fruits.reverse()
print("Reversed fruits list:", fruits)
```

```python
# Q5.Given marks=[70,80,90,60,100]
marks = [70, 80, 90, 60, 100]
# calculate the maximum marks
max_marks = max(marks)
print("Maximum marks:", max_marks)
# calculate the minimum marks
min_marks = min(marks)
print("Minimum marks:", min_marks)
# calculate the average marks
average_marks = sum(marks) / len(marks)
print("Average marks:", average_marks)
```

```python
# Q6. Remove duplicates: x = [1,2,2,3,4,4,5]
x = [1, 2, 2, 3, 4, 4, 5]
unique_x = list(set(x))
print("Original list:", x)
print("List after removing duplicates:", unique_x)
```
